```python
from solver import TemporalSolver
import numpy as np

params = [1.0, 1.0, 0.1]  # [tau, gamma, chi_]
timing = [0, 10, 0.01]

def update_osc(x: np.ndarray, t: float) -> np.ndarray:
    return 1 + 0.2 * np.sin(2*np.pi*t)



solver = TemporalSolver(params, 
                        animate=True,
                        animation_name="chi_time_Ca.gif", 
                        timing=timing, 
                        update_atmospheric=update_osc)
times, alphas = solver.solve()

# solver = TemporalSolver(params, 
#                         animation_name="chi_time_gs.gif", 
#                         timing=timing, 
#                         update_stomata=update_osc)
# solver.solve()

# solver = TemporalSolver(params, 
#                         animation_name="chi_time_K.gif", 
#                         timing=timing, 
#                         update_kappa=update_osc)
# solver.solve()
```

```python


from solver import get_homogeneous_solution

ss_solution = get_homogeneous_solution(0, params)
alpha_ss = params[1] * (1 - ss_solution)

import matplotlib.pyplot as plt
plt.plot(times, alphas)
plt.axhline(alpha_ss, color='red', linestyle='--', label='Steady-state α')
plt.xlabel("Time")
plt.ylabel("Rescaled assimilation rate α")
plt.title("Rescaled assimilation rate over time")
plt.grid()
plt.show()
```
