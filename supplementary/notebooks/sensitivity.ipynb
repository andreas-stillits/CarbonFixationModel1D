```python
%cd /home/andreasstillits/coding/Leaves1D/supplementary
from codebase.utils.plotfunctions import plot_sensitivity_map
import matplotlib.pyplot as plt


def create_nonlinear_sensitivity_figure(
    filename: str = "figures/nonlinear_sensitivity.svg",
) -> None:
    """Create the nonlinear sensitivity figure"""
    mus = ["0.00", "1.00", "8.00"]
    basepath = "files/nonlinear_sensitivities/sensitivities_mu"
    titles = ["linear non-saturating\n"+r"$C < K_m$", "moderately saturating\n"+r"$C \approx K_m$", "strongly saturating\n"+r"$C \gg K_m$"]

    paths = [basepath + mu +"_.txt" for mu in mus]

    _, axs = plt.subplots(1, 3, figsize=(12, 4), sharey=True)
    axs = axs.flatten()
    for ax, path, mu, title in zip(axs, paths, mus, titles):
        im = plot_sensitivity_map(path, ax=ax)
        ax.set_title(title)
        ax.set_xlabel(r"Absorption balance $\tau$ []")
    axs[0].set_ylabel(r"Transport balance $\gamma$ []")
    # put a colorbar to the right
    cbar = plt.colorbar(im, ax=axs, orientation="vertical")
    cbar.set_label(r"Sensitivity $\eta(\tau, \gamma)$ [%]")
    # plt.suptitle("Nonlinear sensitivity maps for increasing levels of saturation")
    plt.savefig(filename)
    plt.savefig(filename.replace(".svg", ".jpg"), dpi=900)
    plt.show()


create_nonlinear_sensitivity_figure()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

x = np.linspace(0, 1)
y = lambda mu: (x - 0.1) / (1 + mu * x)

plt.plot(x, y(0.0), color="black", label=r"$C < K_m$")
plt.plot(x, y(1.0), color="darkblue", linestyle="--", label=r"$C \approx K_m$")
plt.plot(x, y(8.0), color="darkred", linestyle="-.", label=r"$C \gg K_m$")
plt.xlabel("Concentration []")
plt.ylabel("Uptake rate []")
plt.legend()
```
