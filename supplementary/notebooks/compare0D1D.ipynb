```python
import numpy as np 
import matplotlib.pyplot as plt 


N = 100
gammas = np.exp(np.linspace(np.log(0.01), np.log(100), N))
taus = np.exp(np.linspace(np.log(0.01), np.log(100), N))
error = np.ones((N,N))

for i, gamma in enumerate(gammas):
    error[i,:] = 100*gamma*(taus**2/(taus*np.tanh(taus)) - 1)


plt.plot(taus, error[0,:], 'r-', label=r'$\gamma=0.01$')
plt.plot(taus, error[int(N/2),:], 'g-', label=r'$\gamma=1$')
plt.plot(taus, error[-1,:], 'b-', label=r'$\gamma=100$')
plt.xscale('log')
plt.ylim(0, 100)
plt.xlabel(r'$\tau$')
plt.ylabel('Error (%)')
plt.legend()
plt.show()

plt.pcolor(taus, gammas, error, shading="nearest", cmap="inferno", vmin=0, vmax=50)
plt.xscale('log')
plt.yscale('log')
plt.colorbar(label='Error (%)')
plt.xlabel(r'$\tau$')
plt.ylabel(r'$\gamma$')
plt.show()
```
